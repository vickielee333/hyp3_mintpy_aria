```python
import h5py

# Open the .h5 file
file_path = 'timeseries_ERA5_demErr.h5'
with h5py.File(file_path, 'r') as file:
    # Extract the date dataset
    date_data = file['date'][:]

# Convert the byte strings to regular strings
date_list = [date.decode('utf-8') for date in date_data]

# Print the list (optional)
print(date_list)
```

```python
with open('date.dat','w') as file:
    for item in date_list:
        file.write(item+'\n')
```
